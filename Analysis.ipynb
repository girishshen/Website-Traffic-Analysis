    <h2><strong>Step 3 - Feature Engineering</strong></h2>
</div>

```python
feat_df = df.copy()
```

```python
# Average time per page
def compute_avg_time_per_page(row):
    return row['Time on Page'] / row['Page Views'] if row['Page Views'] > 0 else 0

feat_df['Avg Time Per Page'] = feat_df.apply(compute_avg_time_per_page, axis=1)
```

```python
# High bounce indicator (Bounce Rate > 70%)
feat_df['High Bounce'] = (feat_df['Bounce Rate'] > 0.7).astype(int)
```

```python
# Page view intensity: Pages per Second
def compute_pages_per_sec(row):
    return row['Page Views'] / row['Session Duration'] if row['Session Duration'] > 0 else 0

feat_df['Pages Per Sec'] = feat_df.apply(compute_pages_per_sec, axis=1)
```

```python
 # Traffic source one-hot encoding
feat_df = pd.get_dummies(feat_df, columns=['Traffic Source'], prefix='Traffic', drop_first=True)
```

```python
# Loyalty score: log(Previous Visits + 1)
feat_df['Loyalty Score'] = np.log1p(feat_df['Previous Visits'])
```

```python
# Prepare final df for modeling
df = feat_df.copy()
print(f"Dataframe now has {df.shape[1]} columns, including engineered features.")
```

    <h2><strong>Step 5 - Model Selection And Training</strong></h2>
</div>

```python
# Define target and features
df['converted'] = (df['Conversion Rate'] > 0).astype(int)
target_col = 'converted'

X = df.drop(columns=[target_col], errors='ignore')
y = df[target_col]
```

```python
# Identify numerical and categorical columns
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()
```

```python
# Create preprocessing pipelines for both numeric and categorical data
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# Combine transformers into a preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Create the modeling pipeline with RandomForestClassifier
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])
```

```python
# Split the dataset (80% training and 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Predict on the test data
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)
```

```python
# Check Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# Check Classification Report
print(classification_report(y_test, y_pred))
```

```python
unique_classes = np.unique(y_test)
print(f"Unique classes in y_test: {unique_classes}")
```

```python
# Check Confusion Matrix
labels = [0, 1]  # Replace with your actual class labels
cm = confusion_matrix(y_test, y_pred, labels=labels)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Proportion of variance explained by the model (1.0 is perfect) (R2 Score)
r2 = r2_score(y_test, y_pred)
print(f"R2: {r2:.2f}")
```
